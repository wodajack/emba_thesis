Analysis of EnergyConsumption for Freight Trains 2024-2025
data source: rail power box austria



```python
# load packages
import pandas as pd
import numpy as np


# read in 24 excel files with consecutive names, ending with 202401 to 202512 from a specific folder and concatenate them into a single dataframe
# specify the folder path
folder_path = '/Users/Hans/Documents/EMBA/MasterThesis/EnergyTest/Data/consumptiondata/'

# create list of file paths
file_list = [f'{folder_path}AbrechnungEnergie_RCA_{year}{month:02d}.xlsx' 
             for year in [2024, 2025] 
             for month in range(1, 13) if not (year == 2025 and month > 12)]

# read and concatenate all files
df = pd.concat([pd.read_excel(file) for file in file_list], ignore_index=True)
```

```python
# display the first few rows of the combined dataframe
print(df.head())
# check the shape of the combined dataframe
print(df.shape)

# save the combined dataframe to a new csv file in the same folder
output_file = f'{folder_path}combined_consumption_data_2024_2025.csv'
df.to_csv(output_file, index=False)
print(f'Combined data saved to {output_file}')  

# list all column names
print("Column names in the combined dataframe:")
print(df.columns)
```
